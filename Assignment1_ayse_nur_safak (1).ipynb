## ***Coding Part***

```python
import random
def instructions():
    ''' Game Instructions '''
    print('''

    Welcome to Quest for Dragon Egg!

    Your goal is to find the mystic dragon egg, hidden in a faraway clearing.
    To awaken the egg, you need a magic spell from a mage in the tower.
    However, you must navigate through challenges and traps to find it.

    Gameplay overview

    - Move between different Room rooms by typing the name of the room.*
    - Some room may have items or riddles or hidden traps like weathers, so explore them carefully.*
    - Encounter different events, such as a bear asking riddles or finding hidden items.*
    - Manage your energy wisely and collect the necessary items to finish the game.*

  
    Commands
    Moving Between Rooms:

    You can move between these rooms by typing the room name:*
    - Clearing
    - Forest
    - Tower
    - Beekeeper
    - Riverbank
    - Secret Cave

    >Example: To move to the forest, just type forest and press Enter.

    Inventory Check:
    The players can see their inventory.

    >Example:To display your inventory, just type inventory to see the list of items you collected.

    Energy Check:

    You will lose energy every time you make a move. If your energy reaches 0, the game ends.

    Riddle:
    
    The player must solve a riddle to pass the bear without taking damage. 
    To see the riddle the weather must be sunny.

    Energy Restoration:

    Using the potion you need to find first, the player can restore its energy.

    >Example:To use  the potion, just type use potion and press Enter.

    Exiting the Game:

    At any time, you can type exit to quit the game.

    Important Game Mechanics
    Energy Management:

    Each player will start with 100 energy, but each move decreases your energy randomly 
    by 5 to 15 points. If your energy reaches 0, the game is over.
    Keep an eye on your energy by typing energy and plan your moves wisely to conserve it.

    Moves Left:

    Each player has a total of 12 moves to complete the game. 
    Each time you move to a new room, the number of moves decreases.

    Items:

    Each player will need certain items to progress in the game:

    - Honey: Used to distract the bear in the forest. This protects the egg from the bear. 
        If players have no honey or charm, the bear will break the egg and the mission will fail.
    - Dragon Egg: The game's main goal is to find and protect this until the egg is awakened.
    - Amulet: Found at the riverbank, this item is needed to protect the dragon egg from storms 
        and fog which are environmental hazards or bears. Also, if the player enters 
        the secret cave without the charm, they will fall and the egg will break.

    Expected Behavior:
    - Amulet + No Honey: If the player has the amulet but no honey, the charm will still protect 
        the egg from the bear.*
    - No amulet + No Honey: If the player has neither the amulet nor honey, 
        the bear will break the egg.*
    - Honey: If the player has honey, the bear is distracted, and 
        the player can safely pass, regardless of whether they have the charm.

    Random Events:

    The forest and some other rooms have random events, 
    such as riddles or weather changes. Be prepared to adapt!

    How to Win this Game
    - Explore the rooms and solve riddles to find the dragon egg in the clearing.
    - Make sure to visit the riverbank to collect the amulet, which will allow 
        you to awaken the egg safely.
    - Protect the Dragon Egg as you travel through the dangerous forest.
    - If you do not have enough energy, you need to find a potion. 
        By using the potion you can gain 30 energy.
    - Once you have the dragon egg and the charm, go to the tower to get the spell from the mage.
    - Use the spell to awaken the dragon egg, and you win the game!
    - However, there are dangers along the way—such as a bear in the forest and 
        harsh weather conditions that could break the egg if you're not prepared.

        By typing exit, You can finish the game.
    ''')
# Importing Python modules
# Random module allows random selection of events such as weather condition in the forest
import random

# Custom Exceptions
class OutOfEnergyException(Exception):
    '''Raised when the player runs out of energy'''
    pass
#Defining custom expectations to handle specific game-over situations
class GameOverException(Exception):
    ''' Used to end the game when specific conditions are met.'''
    pass

# Decorator to check player energy before actions
#It is used on movvement functions to ensure actions do not proceed  if the player is out of energy
def check_energy(func):
   
    def wrapper(*args, **kwargs):
        if args[0].energy <= 0:
            raise OutOfEnergyException("You have no energy left to continue!")
        return func(*args, **kwargs)
    return wrapper

# Class representing a room in the game
class Room:
    def __init__(self, name, description, interact_func=None):
        '''init method initialise a new room with its name, description and, 
        optional interaction function'''
        self.name = name #name of the room like forest
        self.description = description #description will display when the player enters the room.
        self.interact_func = interact_func  # A specific function to trigger unique events in the room

    def interact(self, player):
        """Trigger the room's interaction function if available"""
        if self.interact_func:
            ''' if any interaction exists for the room, it calls interact_func(player),allowing the room
            to execute its unique events'''
            return self.interact_func(player)
        else:
            #if there is no interaction, it prints description of the room
            #return false if the interaction does not end the game 
            print(self.description)
            return False  # Continue game

# Class representing the player in the game
class Player:
    def __init__(self, name):
        #initialises a new player with their name and starting attributes
        self.name = name #player's name
        self.energy = 100 #player's energy 
        self.moves_left = 12 #total moves the player can make 
        self.inventory = {
            #to track the player's items
            #initialy they set as False, the player needs to collect in the game
            'honey': False, 
            'egg': False,
            'amulet': False,
            'potion': False
        }

    @check_energy #to ensure the player has enough energy 
    #handles the player's movement between rooms and triggers room interactions
    def move(self, room):
        """Move to a new room and interact"""
        #check if moves_left is grater than zero.If not, raises the gameoverException
    
        if self.moves_left <= 0:
            raise GameOverException("You have no moves left!")
        print(f"\n{self.name}, you are moving to the {room.name}.")
        self.moves_left -= 1
        self.energy -= random.randint(5, 15)  # Reduce energy randomly per move
        print(f"Energy: {self.energy}, Moves Left: {self.moves_left}")#prints the player's movement and decreased energy
        return room.interact(self)#calls the interact method of the room to trigger any events for any specific rooms

    def show_inventory(self):
        """Display player's current inventory"""
        print("\nInventory:")
        for item, has_item in self.inventory.items():
            if has_item:#if any item marked as True, it prints the item.
                print(f"- {item}")
        if not any(self.inventory.values()):#if there is no item,it prints the message
            print("- Your inventory is empty.")

    def break_egg(self, cause):
        #manges the events where the egg is broken
        #check if the egg is in the player's inventory.
        #if there is an egg, prints a failure message and returns True to end the game
        #if the player does not have the egg, it returns false.
        """Handles the event of the dragon egg breaking."""
        if self.inventory['egg']:
            print(f"\nUnfortunately, the {cause} caused the dragon egg to break. You have failed the quest.")
            return True  # Game over
        return False

    def use_potion(self):
        """Use a potion to restore energy"""
        #Checks if the player has a potion in inventory.
        #If the player want to use the potion, it needs to write use potion
        #the potion restores 30 energy and set potion as False in the inventory.
        #if the player has not any potion, it will print you do not have any potions messages.
        if self.inventory['potion']:
            print("You drink the potion and restore 30 energy!")
            self.energy += 30
            self.inventory['potion'] = False  # Use up potion
        else:
            print("You don't have any potions!")

# Game interactions
def forest_event(player):
    """Handle random events in the forest"""
    #Weather 
    #if it is stormy and player has not amulet, it will break the egg.
    #If the egg is broken, brake_egg method will be called.
    #Fog without an amulet disorients the player but does not end the game.
    
    weather = random.choice(["sunny", "stormy", "foggy"])
    print(f"The weather in the forest is {weather}.")

    # Protection against storm if the player has the amulet
    if weather == "stormy" and player.inventory['egg']:
        if player.inventory['amulet']:
            print("The storm rages, but your amulet protects the dragon egg.")
        else:
            return player.break_egg("storm")  # End game if no amulet

    # Foggy weather now only causes disorientation if no amulet, but no damage
    if weather == "foggy" and not player.inventory['amulet']:
        print("The fog is thick and disorienting! You proceed cautiously.")
    #if the player has not the egg, the storm blocks u. but the weather is foggy, u can move   
    if weather == "stormy":
        print("The storm blocks your way further into the forest.")
        return False
    elif weather == "foggy":
        print("The fog is thick, but you can proceed cautiously.")

    # Protection against bear if the player has the amulet or honey
    if not player.inventory['honey'] and player.inventory['egg']:
        if player.inventory['amulet']:
            print("The bear growls angrily, but your amulet emits a protective glow, and the bear backs off.")
            return False  # The amulet saves the egg from the bear
        else:
            #The bear breaks the egg if the player lacks honey and the amulet.
            print("The bear growls angrily and attacks you! Without honey or the amulet to distract the bear, it breaks the dragon egg!")
            return player.break_egg("bear")
    elif not player.inventory['honey']:
        events = [
            #If neither honey nor the amulet is available, the bear blocks the way, and the player must solve a riddle to proceed.
            #This lambda function simply returns False when called. It represents a scenario where the bear blocks the player's way without requiring further action.
            ("The bear growls and blocks your way.", lambda: False),
            #This lambda function calls the solve_riddle(player) function when triggered, which prompts the player to solve a riddle posed by the bear.
            ("The bear asks you a riddle to let you pass.", lambda: solve_riddle(player))
            
        ]
        event_text, event_action = random.choice(events)
        print(event_text)
        return event_action()
    else:
        print("You give the honey to the bear. The bear lets you pass.")
        return False

def solve_riddle(player):
    """Solve a riddle to pass the bear"""
    riddle_answer = input("Riddle: You can remove its skin and it won’t cry- but you may!").lower()
    if riddle_answer == "onion":
        print("Correct! You may pass.")
        return False
    else:
        print("Wrong answer! The bear blocks your way.")
        return False  # No damage now, but player must try again or find honey

def tower_event(player):
    """Interact with the tower - win only if the egg is found"""
    # Checks for the dragon egg in the player’s inventory to end the game successfully.
    if player.inventory['egg']:
        print("The mage gives you the spell to awaken the dragon egg. Quest complete!")
        return True  # End game only if the egg is found
    else:
        #Otherwise, the mage informs the player they still need to find the egg.
        print("You haven't found the dragon egg yet. The mage can't help you.")
    return False

def beekeeper_event(player):
    #Allows the player to obtain honey from the beekeeper, which distracts the bear.
    """Interact with the beekeeper"""
    if not player.inventory['honey']:
        #If the player doesn’t have honey, they receive it.
        print("The beekeeper gives you some honey to help with the bear.")
        player.inventory['honey'] = True
    else:
        # If they already have it, nothing happens.
        print("You already have honey.")
    return False

def riverbank_event(player):
    """Interact with the riverbank"""
    #Provides the amulet, which protects the egg from certain events.
    if not player.inventory['amulet']:
        #If the player doesn’t have the amulet, they obtain it. 
        print("You found a mysterious amulet. It seems to glow with power.")
        player.inventory['amulet'] = True
    else:#Otherwise, it does nothing.
        print("You've already collected the amulet.")
    return False

def secret_cave_event(player):
    """Hidden room interaction - the egg breaks if you enter without the amulet"""
    if player.inventory['amulet']:
        #With the amulet, the player finds a potion in the cave.
        print("The amulet reveals a hidden cave entrance!")
        if not player.inventory['potion']:
            print("You find a potion inside the cave!")
            player.inventory['potion'] = True
        else:#If the player already has potion
            print("You've already collected the potion.")
    else:
        #Without the amulet, the player risks breaking the egg if they have it.
        print("Without the amulet, you trip and fall as you enter the cave.")
        if player.inventory['egg']:
            return player.break_egg("falling in the cave")  # Break egg if no amulet
        print("You proceed carefully but without the egg, nothing breaks.")
    return False

def clearing_event(player):
    """Interact with the clearing to find the dragon egg"""
    if not player.inventory['egg']:
        print("You found the dragon egg in the clearing! It's a majestic and powerful artifact.")
        player.inventory['egg'] = True
    else:
        print("You are back in the clearing, but you already have the dragon egg.")
    return False

# Function to initialize the game world
def initialize_game():
    # Define rooms
    clearing = Room("clearing", "You are in a peaceful clearing, and something seems hidden here.", clearing_event)
    forest = Room("forest", "You are in a dense forest with a bear blocking your way.", forest_event)
    tower = Room("tower", "You have reached the mage's tower. The mage awaits your return.", tower_event)
    beekeeper = Room("beekeeper's house", "You are at the beekeeper's house.", beekeeper_event)
    riverbank = Room("riverbank", "You are at a quiet riverbank, and something catches your eye.", riverbank_event)
    secret_cave = Room("secret cave", "A hidden cave lies here.", secret_cave_event)  # Hidden room

    return {
        'clearing': clearing,
        'forest': forest,
        'tower': tower,
        'beekeeper': beekeeper,
        'riverbank': riverbank,
        'secret cave': secret_cave  
    }

# Main game loop
def play_game():
    #printing instructions for playing game
    print("Welcome to the Quest for the Dragon Egg!")
    instructions() 
    # Create player
    player_name = input("Enter your name, brave adventurer: ")
    player = Player(player_name)

    # Initialize room
    rooms = initialize_game()

    current_room = rooms['clearing']
    game_over = False

    while not game_over:
        try:
            next_room_name = input("\nWhere would you like to go? (clearing, forest, tower, beekeeper, riverbank, secret cave): ").lower()
            if next_room_name == 'exit':#by typing exit u can finish the game
                print("Thank you for playing!")
                break
            elif next_room_name == 'inventory':#inventory check
                player.show_inventory()
            elif next_room_name == 'use potion':#for using potion
                player.use_potion()
            elif next_room_name in rooms:
                current_room = rooms[next_room_name]
                game_over = player.move(current_room)
            else:#Ends the game if energy is depleted or the egg breaks.
                print("Invalid room. Please choose a valid location.")
        except OutOfEnergyException:
            print("You are too exhausted to continue. Game Over!")
            break
        except GameOverException as e:
            print(e)
            break

# Start the game
if __name__ == "__main__":
    play_game()
```
